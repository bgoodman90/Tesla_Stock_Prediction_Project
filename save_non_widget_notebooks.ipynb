```python
import nbformat

# Load the notebook
nb_path = "tesla_sentiment_analysis.ipynb"
nb = nbformat.read(nb_path, as_version=nbformat.NO_CONVERT)

# Remove all 'widgets' metadata at every level
for cell in nb.cells:
    if 'metadata' in cell:
        cell.metadata.pop('widgets', None)
    # Also check outputs (sometimes widgets are stored there)
    if 'outputs' in cell:
        for output in cell['outputs']:
            if 'metadata' in output:
                output['metadata'].pop('widgets', None)

# Save the cleaned notebook
nb_clean_path = "tesla_sentiment_analysis_clean.ipynb"
nbformat.write(nb, nb_clean_path)
from nbformat import read, write, NO_CONVERT

nb = read(nb_clean_path, as_version=NO_CONVERT)
for cell in nb.cells:
    if 'outputs' in cell:
        cell['outputs'] = []

write(nb, nb_clean_path)
print(f"Outputs cleared: {nb_clean_path}")

print(f"Cleaned notebook saved as {nb_clean_path}")
```

```python
import nbformat

# Load your notebook
notebook_path = "tesla_sentiment_analysis.ipynb"
nb = nbformat.read(notebook_path, as_version=4)

# -----------------------------
# Step 1: Inspect for any widgets metadata
# -----------------------------
widgets_found = False
for cell in nb.cells:
    if "metadata" in cell and "widgets" in cell.metadata:
        widgets_found = True
        print("Widgets metadata found in a cell:", cell.metadata['widgets'])

if not widgets_found:
    print("No widget metadata found in any cells.")

# -----------------------------
# Step 2: Option 1 â€” Remove all widgets metadata safely
# -----------------------------
for cell in nb.cells:
    if "metadata" in cell and "widgets" in cell.metadata:
        del cell.metadata["widgets"]

# -----------------------------
# Step 3: Save cleaned notebook
# -----------------------------
cleaned_path = "tesla_sentiment_analysis_cleaned.ipynb"
nbformat.write(nb, cleaned_path)
print(f"Cleaned notebook saved to {cleaned_path}")
```

```python
import json

# Path to your notebook
notebook_path = "tesla_sentiment_analysis.ipynb"
cleaned_path = "tesla_sentiment_analysis_cleaned.ipynb"

# Load notebook JSON
with open(notebook_path, 'r', encoding='utf-8') as f:
    nb_data = json.load(f)

# Recursive function to remove all "widgets" keys
def remove_widgets(obj):
    if isinstance(obj, dict):
        if "widgets" in obj.get("metadata", {}):
            del obj["metadata"]["widgets"]
        for k, v in obj.items():
            remove_widgets(v)
    elif isinstance(obj, list):
        for item in obj:
            remove_widgets(item)

# Clean notebook
remove_widgets(nb_data)

# Save cleaned notebook
with open(cleaned_path, 'w', encoding='utf-8') as f:
    json.dump(nb_data, f, indent=1)

print(f"Cleaned notebook saved to {cleaned_path}")
```
